```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation
from keras.utils import to_categorical
from keras.optimizers import SGD
```

```python
X_train = open('train_images.txt').readlines()[0]
X_train = [int(x.strip()) for x in X_train.split()]
X_train = np.array(X_train).reshape(-1, 28, 28, 1)

X_test = open('test_images.txt').readlines()[0]
X_test = [int(x.strip()) for x in X_test.split()]
X_test = np.array(X_test).reshape(-1, 28, 28, 1)

Y_train = open('train_labels.txt').readlines()[0]
Y_train = [int(x.strip()) for x in Y_train.split()]
Y_train = to_categorical(np.array(Y_train), num_classes=10)

data = open('test_labels.txt', encoding='utf-8').readlines()[0]
Y_test = []
for x in data.split():
    try:
        Y_test.append(int(x.strip()))
    except:
        Y_test.append(7)

Y_test = to_categorical(np.array(Y_test), num_classes=10)
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
from scipy.misc import imsave

X = X_train.reshape(-1, 28, 28)
for i in range(100):
    imsave('img/images_{}.jpg'.format(i), X[i, :, :])
```

```python
#Densly connected neural network
dnn = Sequential()

dnn.add(Dense(100, input_shape=(784,), activation='relu'))
dnn.add(Dropout(0.4))

dnn.add(Dense(50, activation='relu'))
dnn.add(Dropout(0.4))

dnn.add(Dense(10))
dnn.add(Activation(activation="softmax"))

dnn.summary()
```

```python
batch = 500
eps = 50

opt = SGD(lr=0.01, decay=1e-5)
dnn.compile(opt, loss='categorical_crossentropy', metrics=['accuracy'])

dnn.fit(X_train.reshape(-1, 784), Y_train, batch_size=batch, epochs=eps, validation_data=(X_test.reshape(-1, 784), Y_test))
```

```python
#The Model
model = Sequential()

model.add(Conv2D(32, (3, 3), padding='SAME', activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (5, 5), padding='SAME', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(100, activation='relu'))
model.add(Dropout(0.4))

model.add(Dense(10))
model.add(Activation(activation="softmax"))

model.summary()
```

```python
batch = 500
eps = 50

opt = SGD(lr=0.01, decay=1e-5)
model.compile(opt, loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, Y_train, batch_size=batch, epochs=eps, validation_data=(X_test, Y_test))
```
